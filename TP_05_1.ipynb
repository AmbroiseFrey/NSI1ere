# <span style="color:red;"><center>TP - Créer une page web sans CSS </center></span>

## 1. Objectif du TP

Créer une page web simple, sans CSS, comportant des liens, des images, ... et une structure de base *header*, *section*, *footer*.

## 2. Le TP de Eric Hartung

Un collègue ayant déjà eu la gentillesse de mettre en ligne un TP complet à ce sujet, vous allez simplement réaliser le travail demandé à l'adresse suivante :

[TP : Ma première page web](http://mathartung.xyz/isn/TP/TPhtml/tp1.html)

**Sauf pour le dépot de votre travail bien sûr.** Celui-ci sera à déposer sur **Atrium** (seulement le fichier HTML) dans le **casier "Page sans CSS"**.
