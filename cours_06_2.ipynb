## B. Architecture réseau

### I. Fonctionnement des réseaux informatiques

#### 1. Qu'est-ce qu'un réseau informatique ?

De manière générale, on appelle **réseau** un **ensemble d'entités liées entre elles**, c'est-à-dire relier par des liens qui leur permettent d'échanger de l'information entre elles. Dans les réseaux sociaux, les entités qui communiquent sont des humains. Dans un **réseau informatique**, ce sont des **machines (ordinateurs) qui communiquent entre elles**.

**Internet** est un **réseau informatique mondial** constitué d'une **multitude d'autres réseaux** de plus en plus petits jusqu'à arriver à votre propre **réseau local** constitué des machines que vous avez chez vous et qui sont reliées à internet par l'intermédiaire de votre box.

**Internet** s'est **construit petit à petit** en agrégeant les premiers réseaux historiques, comme *Arpanet* (1970) aux Etats-Unis et *Cyclades* (1971) en France. Mais pour que des machines du monde entier, aussi différentes soient-elles, puissent communiquer entre elles, il a fallu mettre au point des **protocoles de communication que partagent toutes les machines connectées à Internet**.

De la même façon que les humains du monde entier ont besoin d'une langue commune s'ils veulent communiquer entre eux.

#### 2. Les protocoles de communication

**Sans protocole de communication pré-établi, il est extrêmement compliqué de communiquer.** Si un jour nous sommes amenés à rencontrer d'autres espèces intelligentes, on peut difficilement imaginer à quel point l'absence de protocoles communs risque de compliquer la prise de contact.

Sur Terre, **nous partageons tous à peu près le même protocole de communication orale** (à quelques variantes près en fonction de la situation) :
- **Ouverture d'une communication** par échange de regards et/ou salutations.
- **Echange d'informations** (questions/réponses).
- **Fermeture de la communication** (formules de politesses, signes de la main).

Si vous ne respectez pas ce protocole, notamment son ordre, il y a fort à parier que vos interlocuteurs apprécient peu la conversation avec vous.

Il en va de même pour les **ordinateurs sur Internet**, sauf qu'eux ils **communiquent à distance** et qu'ils sont des **milliards à communiquer en même temps** en utilisant le **même réseau** (un peu comme si on parlait tous en même temps dans une même pièce).

Par conséquent, **leurs protocoles de communication** ont besoin d'un peu **plus de sophistication**. Mais comme pour les humains ce sont rien d'autres que des algorithmes de communication du type *si ceci reçu alors renvoyer ça* ou encore *tant que pas reçu ça continuer à envoyer ceci*.

#### 3. Les protocoles sur Internet

Pour fonctionner, **les protocoles de communication sur Internet doivent** notamment :
- **être indépendants du matériel** (si j'ai un ordinateur DELL récent, je dois pouvoir communiquer sans problème avec un vieil HP),
- **proposer une identification unique de l'émetteur et du récepteur** (pour permettre d'acheminer les messages au bon endroit),
- **être capables de gérer les flux de données** (pertes, encombrement, débit, ...)
- **travailler avec de l'information exclusivement binaire** (des 0 et des 1 ... on parle d'ordinateurs ...)

C'est l'**invention du protocole TCP** (1974) **suivi de près du protocole IP**, les deux travaillant en très étroite collaboration, qui a **permis d'interconnecter les premiers réseaux** qui s'étaient développés indépendamment les uns des autres. Même s'ils ont évolué depuis, le couple de protocoles TCP/IP est **aujourd'hui encore la colonne vertébrale d'Internet**.

#### 4. Le modèle en couches

Sur Internet, ce sont en fait des **applications (des logiciels) qui communiquent**. Si vous surfez sur le web, par exemple, c'est votre navigateur internet qui se charge d'envoyer vos demandes sur le net et ensuite de récupérer les réponses.

Mais si chaque logiciel, se charge directement de cette communication, cela signifie que le jour où les protocoles de communications changent, il faut réécrire tous les logiciels qui communiquent sur Internet. Ce serait ingérable.

C'est une des raisons qui a conduit à mettre en place ce qu'on appelle le **modèle en couches**. Dans ce modèle, la communication est divisée en **différents intermédiaires**, situées les uns au-dessus (ou en-dessous) des autres et **qui gèrent chacun un niveau de communication**.

**Une analogie postale :**

Le modèle en couches relève un peu de la **même logique que les services postaux**.

Si vous voulez envoyer une lettre à grand Mamie qui habite en Bretagne, rien ne vous empêche de traverser la France pour la lui remettre en mains propres. Mais il est quand même plus simple de passer par la Poste, qui va prendre votre courrier et le tamponner pour en indiquer la validité. Puis la confier à un service de transports (routier par exemple) qui va l'acheminer au centre de tri le plus proche de chez vous et la mettre dans un avion à destination du centre de tri le plus proche de la destination. Une fois là-bas, le centre de tri la fera parvenir au bureau de poste local qui se chargera ensuite de sa distribution.

Dans ce système, le seul interlocuteur de l'émetteur (vous) et du récepteur (grand Mamie) c'est la Poste. Et comme ni l'un, ni l'autre n'a besoin de savoir si la lettre est acheminée par avion ou par train ou par quel chemin elle est passée, cela ne fait aucune différence pour vous ou elle en cas de changement.

### II. Le modèle en couches TCP/IP

#### 1. Principe général


**Remarque :** La couche *Réseau* est aussi appelée couche *Internet* et la couche *Accès au Réseau* peut parfois aussi être appelée, par abus de langage, couche *Physique*. En réalité, dans le modèle en couches OSI plus complet que le modèle TCP/IP (il comporte 7 couches au lieu de 4), la couche accès réseau est séparée en 2 couches distinctes, la couche *Physique* (tout en bas) et la couche *Liaison de données* (juste au dessus).

Si on poursuit l'**analogie postale** précédente :
- La couche *Application*, c'est **vous et grand Mamie**.
- La couche *Transport*, ce sont les **bureaux de poste** à côté de chez vous et de grand Mamie.
- La couche *Réseau*, ce sont les **centres de tri**.
- La couche *Accès au Réseau*, se sont les **transporteurs** (avion, train, camion) qui vont d'un centre de tri à l'autre.



**Chaque couche ne communique** qu'avec la couche **immédiatement au-dessus ou en dessous d'elle**, en utilisant des **protocoles qui lui sont propres**.

#### 2. La couche Application

**Toutes les appications utilisateurs qui communiquent sur Internet sont dans cette couche.** On y trouve bien sûr les **navigateurs internet** qui utilisent le protocole *HTTP* (voir chapitre précédent) mais aussi les **gestionnaires de mail** qui utilisent *SMTP* (Simple Mail Transfert Protocol) ou encore les **logiciels de téléchargement** qui utilisent *FTP* (File Transfert Protocol).

A chaque fois qu'elles ont besoin d'une ressource sur Internet, ces applications s'adressent à la couche Transport pour lui indiquer ce dont elles ont besoin et attendent ensuite sa réponse. **Cette couche ne communique qu'avec la couche Transport.**

#### 3. La couche Transport

Les **deux principaux protocoles** de cette couche sont *TCP* (Transmission Control Protocol) et *UDP* (User Datagram Protocol).

Ils vont s'occuper de **mettre en forme les données reçues de la couche Application pour qu'elles puissent voyager efficacement sur Internet** (un peu comme si c'était la poste qui s'occupait de taper votre lettre et de la mettre dans l'enveloppe).

En particulier, la taille des paquets qui peuvent être acheminés sur Internet étant très petite, ils s'occupent de **découper les données reçues en paquets à la bonne taille**.

La différence principale entre *TCP* et *UDP* est que *TCP* **s'assure de la bonne réception des paquets à destination et de leur remise dans l'ordre** (comme ils ne passent pas tous par le même chemin, le premier parti n'arrive pas forcément le premier) alors que *UDP* non. On dit que **TCP est fiable mais pas UDP**.

En contrepartie, *UDP* **est plus rapide que** *TCP* (puisque plus simple), ce qui en fait un protocole de choix pour le **streaming de vidéos** par exemple.

#### 4. La couche Internet ou couche Réseau

**La couche Internet est le domaine du protocole** *IP* (Internet Protocol).

Celui-ci **prend en charge les paquets reçus de la couche Tranport** pour en faire des paquets IP capables de voyager sur le réseau. Pour cela, il utilise notamment le **système des adresses IP** et des **routeurs** munis de **protocoles de routage** qui leur permettent de savoir vers où envoyer les paquets IP reçus.

Avec toujours la même analogie, votre bureau de poste n'a pas vraiment besoin de connaître la destination de votre lettre puisque de toute façon il va la transmettre au centre de tri le plus proche. Par contre, le centre de tri doit être capable de décider où envoyer la lettre en fonction de sa destination.

#### 5. La couche Accès au Réseau

C'est dans cette couche que les **paquets IP sont transformés en signaux physiques** (ondes électromagnétique pour le WIFI et le Bluetooth ou sigaux électriques pour l'Ethernet ou encore signaux optiques pour la Fibre).

Dans cette couche, ce ne sont **pas les adresses IP qui sont utilisées mais les adresses MAC** (aussi nommées **adresses physiques***) qui **correspondent de façon unique aux interfaces réseaux** (cartes réseaux). Ainsi, il y a souvent **plusieurs adresses MAC pour une même adresse IP** (une pour la carte WIFI, une pour la carte Ethernet et une pour la carte Bluetooth par exemple).

### III. Les protocoles TCP/IP

#### 1. L'encapsulation

Le **modèle en couche** présenté dans la partie précédente **repose en partie sur le principe d'encapsulation**. Là encore, l'analogie postale fonctionne bien (comment feront les profs de NSI quand les lettres papier n'existeront plus ?).

Lorsque vous **envoyez une lettre**, vous la mettez **dans une envelopp**e dont l'intérêt est double. D'une part elle permet de **cacher l'information contenue dans la lettre** qui ne concerne que vous et votre destinataire. Et d'autre part, elle vous permet d'**inscrire l'adresse du destinataire ainsi que l'adresse de l'expéditeur** (la vôtre) afin de pouvoir recevoir une réponse ou un accusé de réception.



**Mettre une lettre dans une enveloppe, en informatique c'est faire de l'encapsulation**.

Dans le modèle en couche, de façon imagée, **à chaque fois qu'une couche passe l'information à une couche située en dessous d'elle, elle la met dans une enveloppe**. Cela permet de "cacher" ce qui ne concerne pas la couche du dessous, tout en ne laissant apparaître que ce que lui est utile.


 A l'inverse, **à chaque fois qu'une couche reçoit de l'information venant d'une couche du dessous, elle commence par ouvrir l'enveloppe** pour voir les informations qui lui sont destinées, c'est la **désencapsulation**.


Une fois de plus, c'est un peu **pareil pour le courrier**. Une fois que vous avez mis la lettre dans l'enveloppe, le **bureau de poste va lui apposer son cachet**, qui est une information ajoutée et nécessaire pour que la lettre soit ensuite traitée par le centre de tri.

Le **centre de tri va ensuite rassembler dans des boites**, portant le nom du centre de tri destinataire, toutes les lettres à destination d'un même centre de tri.

Puis, le **transporteur va à son tour rassembler dans un container** tout ce qui va au même endroit. Et **à l'arrivée, on re-déballe tout dans l'autre sens**. (Enfin je crois, je suis pas allé vérifier !)

#### 2. Le protocole IP

Voilà ce que **encapsuler** ("mettre dans une enveloppe") signifie **pour le protocole IP, lorsqu'il reçoit de la part de TCP de l'information à transmettre** (ici "Comment ça va ?"):




Le paquet de données ainsi crée est appelé **paquet IP** (on parle parfois aussi de datagramme). Un paquet IP devient une **trame lorsqu'il est à son tour encapsulé par la couche inférieure** (*Accès au Réseau*).

Toutes les **informations nécessaires l'acheminement d'un paquet IP sont dans son entête** alors que l'**information à transmettre** (les *data*) est "cachée" **dans le corps** du paquet.

#### 3. Les adresses IP

De même qu'il faut que votre adresse postale soit unique pour que votre courrier arrive bien chez, il **faut aussi que les adresses IP soient uniques au monde**.

Une **adresse postale classique** se présente sous la forme : *Nom - Numéro - Rue - Ville - Pays*
Plusieurs personnes peuvent habiter dans la même rue ou au même numéro, et même avoir le même nom. Mais il n'y a qu'**une seule personne dont l'adresse comporte tout ça à la fois**.

Une **adresse IP est contruite à peu près sur le même principe**. Elle est formée (IPV4) de **4 nombres binaires codés chacun sur un octet** (dont avec des valeurs décimales allant de 0 à 255) et séparés par des ".", par exemple : 192.168.0.42 (ou plutôt 11000000.10101000.00000000.00101010 mais c'est moins pratique)

**Remarque :**

En **IPV4** (protocole IP version 4), on peut contruire $256^4=2^{32}\approx 4,3 \cdot 10^9$, soit **un peu plus de 4 milliards d'adresses différentes** or il y a d'ores et déjà bien plus que ça d'objets connectés à Internet (entre 30 et 40 milliards en 2021).

C'est pourquoi nous sommes **progressivement en train de passer à l'IPV6** (protocole IP version 6) où une **adresse IP est composée de 8 nombres binaires codés chacun sur deux octets** (généralement écrits en hexadécimal à 4 chiffres, soit par exemple 2001:0db8:0000:85a3:0000:0000:ac1f:8001).

C'est ainsi qu'avec une adresse codée au total sur 16 octets, soit **128 bits**, nous disposons de $2^{128} \approx 3,4 \cdot 10^{38}$ ou encore **340 sextillions adresses uniques**. Ce qui devrait être largement suffisant, même pour conquérir l'Univers tout entier.

Le plus souvent, les **3 premiers nombres de l'adresse IP permettent d'identifier le sous-réseau** auquel appartient la machine (le pays, la ville, la rue et le numéro si vous voulez) alors que le **dernier correspond au numéro de la machine dans ce sous-réseau** (le nom de la personne du coup).

En pratique, cette **distinction est importante**. En effet, vous n'avez besoin de rien, ni de personne pour aller demander du sel à votre voisin de palier mais c'est un autre histoire si l'envie vous prend de solliciter quelqu'un dans une autre ville. De même, **des machines situées dans le même sous-réseau peuvent communiquer sans problème** mais elles devront demander de l'**aide à un routeur si elles ne sont pas sur le même sous-réseau**.

**Remarque :**

Pour **indiquer où s'arrête la partie de l'adresse IP désignant le réseau et donc où commence celle désignant la machine dans ce sous-réseau**, on utilise les **masques de sous-réseau**.

Par exemple, IP: 192.168.1.6 avec le masque 255.255.0.0 indique la machine numéro 1.6 du réseau 192.168.0.0. Ce n'est pas la même machine que IP: 192.168.1.6 avec le masque 255.255.255.0 qui est la machine 6 du réseau 192.168.1.0.

Ainsi, les masques de sous-réseau permettent d'une certaine façon de **démultiplier les adresses IP** puisque **2 machines peuvent avoir la même adresse IP si leurs masques de sous-réseau sont différents**.

Ce principe est notamment utilisé pour les **réseaux locaux situés derrière les box internet**.

#### 5. Le protocole TCP

Comme on l'a vu, le protocole IP s'occupe d'essayer d'amener à destination les paquets qui lui sont confiés en trouvant le meilleur chemin pour ça. Mais son travail s'arrête là. En particulier, **IP n'offre aucune garantie que le paquet est bien arrivé à destination**. Si la **fiabilité est importante**, il faut alors ajouter un service d'**accusé de réception**, et ça, c'est le **travail de TCP**.

Si vous voulez envoyer un **courrier important**, vous l'envoyez en **Recommandé avec accusé de réception**. Cela signifie qu'au lieu de juste déposer la lettre dans la boite du destinataire en supposant qu'il la trouvera, le facteur doit **donner la lettre en main propre** et **faire signer le destinataire**, puis vous **renvoyer un accusé de réception pour vous confirmer que la lettre est arrivée à la bonne personne**. C'est **exactement ce que fait TCP**.

Concrètement, **après avoir découpé les données à transmettre en petits paquets de la bonne taille**, TCP les fournit à IP en :
- **Numérotant les paquets** envoyés.
- **Vérifiant que le destinataire est prêt à recevoir** des données.
- **Emettant des accusés de réception** quand un paquet arrive à destination.
- **Remettant les paquets dans l'ordre à l'arrivée** (TCP est au départ et à l'arrivée bien sûr).
- **Renvoyant les paquets** dont les accusés de réception mettent trop de temps à arriver.

**Remarque :**

Pour savoir **à quelle application des machines de départ et d'arrivée les paquets reçus sont destinés**, TCP utilise la notion de **ports**. Les ports permettent d'**identifier les applications concernées**, un peu comme l'adresse IP permet d'identifier la machine.

Dans l'analogie postale, le port correspondrait au **nom précis de la personne concernée pour un foyer comportant plusieurs personnes**.

En pratique, le port correpond plutôt à un type de **protocole applicatif**. Par exemple, le port 80 correspond au protocole HTTP, le port 443 au protocole HTTPS et le port 25 au protocole SMTP.

### IV. Relier physiquement les ordinateurs entre eux

Tout ça, c'est bien joli, mais concrètement, pour relier des ordinateurs en réseau, on fait comment ?

#### 1. La connexion réseau

La **connexion physique** au réseau peut se faire en **mode filaire** en utilisant des **câbles Ethernet** munis de **prises RJ45**.


Mais aussi en **mode sans-fil** en passant par le **Wi-Fi** (*Wireless Fidelity*).

Pour utiliser le **mode filaire**, la machine doit être **munie d'un port RJ45 femelle** monté sur une **carte réseau** ou directement sur la **carte mère**.


#### 2. Relier plus de 2 machines

Le matériel précédent permet de relier directement deux machines entre-elles mais **au delà de deux machines**, il faut utiliser un **switch**.


Le switch est l'équivalent d'une **multiprise**, il se contente de faire transiter l'information de n'importe laquelle de ses ports vers tous ses autres ports. Il est donc transparent sur le réseau et n'a **pas d'adresse IP**. Il en existe de différentes tailles (de 8 à 24 ports sur la photo).

#### 3. Les adresses MAC

Dans les **câbles Ethernet circulent des trames Ethernet** (trames Wi-Fi si connexion en Wi-Fi mais c'est globalement la même chose).

Comme on l'a vu dans le **modèle en couches**, ces trames sont obtenues par **encapsulations successives** :


Ainsi, du fait de l'encapsulation, l'**adresse IP n'est pas visible** au niveau de cette couche *Accès au Réseau* et c'est une **autre adresse** qui est utilisée pour diriger l'information, l'**adresse MAC** (*Media Access Control*) ou **adresse physique**.

L'adresse MAC n'est **pas associée à une machine mais à une carte réseau**. Elle est **codée sur 6 octets** et on l'écrit en général sous la forme de **6 nombres hexadécimaux à 2 chiffres séparés par ":"** (*00:E0:4C:68:02:11* par exemple).

Les adresses MAC sont **attribuées de manière unique à une carte réseau par son constructeur, au moment de sa fabrication**. D'ailleurs, les 3 premiers octets servent à désigner le constructeur de la carte (*00:E0:4C* pour Realtek par exemple).

#### 4. Protocole ARP

Comme toutes les autres couches, la couche *Accès Réseau* a son propre protocole, le **protocole ARP** (*Adress Resolution Protocol*). Son rôle est essentiellement de trouver l'adresse MAC de destination.

Pour cela, il envoie une requête *broadcast* **à toutes les machines du sous-réseau** du genre : "Quelle est l'adresse MAC de la machine ayant l'adresse IP suivante ?". Une fois l'adresse MAC obtenue, il **encapsule le paquet IP dans une trame ethernet sur laquelle n'apparaît plus que l'adresse MAC correspondante**.

#### 5. Passerelle et routeur

Une passerelle (*gateway*) est un **équipement permettant de relier ensemble deux réseaux**. Un switch est donc déjà une passerelle mais le plus souvent, on utilise ce terme lorsque l'autre réseau est Internet (on parle parfois de passerelle internet).

Sur votre **réseau local domestique** (à la maison), c'est en général la **box internet fournie par votre FAI** (Orange, Free, SFR, ...) qui a ce **rôle de passerelle entre votre réseau local et Internet** (on parle parfois de passerelle internet). 

Ce type de passerelle fait aussi de votre box un **routeur**, mais ce n'est pas tout car celle-ci joue aussi le rôle de **serveur DNS** et de **serveur DHCP**.

#### 6. Serveur DNS et DHCP

Ce n'est pas pour rien que les box sont dites "tout-en-un". En plus de vous permettre de relier votre réseau local (et donc toutes les machines qui y sont connectées) à Internet, elle assure aussi le rôle de **serveur DNS** (*Domain Name Server*) et **serveur DHCP** (*Dynamic Host Configuration Protocol*).

Grâce au **protocole DHCP**, c'est elle (en général, on peut faire autrement) qui **attribue une adresse IP locale** (du genre 192.168.1.42) **à tous vos équipements** (ordinateurs, tablettes, smartphones, TV, console, ...) **connectés à votre réseau local**, que ce soit en Ethernet ou en Wi-Fi.

Grâce au **protocole DNS**, elle vous permet de taper *www.google.fr* au lieu de *216.58.208.227* dans votre navigateur, ce qui est quand même plus pratique. Ce protocole permet en effet d'**associer une IP aux différents noms de dommaine enregistrés sur le Web**.
