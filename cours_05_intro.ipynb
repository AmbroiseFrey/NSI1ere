## Introduction

Une des spécificités de l'Homme moderne est sans doute sa capacité à **fabriquer des machines**, c'est-à-dire des objets capables d'accomplir diverses tâches, telle que demandées par un utilisateur humain.

Pour cela, la machine doit disposer d'une **interface** qui permet à l'utilisateur de la **faire fonctionner**, de la piloter. Et dès que la machine est un tant soit peu perfectionnée, cette interface permet aussi à la machine de **renvoyer des informations** sur son fonctionnement ou le résultat de son travail vers l'utilisateur. On parle alors d'**interface homme-machine (IHM)**.

Par exemple, une **voiture** est munie d'un volant, de pédales, de boutons, ... qui permettent de la conduire mais aussi de cadrans, de voyants et d'écrans qui informent sur son fonctionnement. Tout cela forme l'interface homme-machine de la voiture.

Evidemment les **ordinateurs** sont aussi des machines et possèdent donc aussi des IHM. La plus répandue est l'ensemble des **périphériques d'entrée et de sortie** {clavier, souris, écran, enceintes, imprimante}. Mais si on va plus loin, les **logiciels** font aussi partie de cette IHM puisqu'ils permettent à l'utilisateur et à l'ordinateur de communiquer afin que ce dernier accomplisse le travail souhaité.

En particulier, lors de la **navigation sur le Web**, les internautes interagissent avec leur machine par le biais des pages Web. Dans ce cas, l'interaction repose sur des langages spécifiques que sont le *HTML*, le *CSS*, le *Javascript* ou encore le *PHP*.

#### 2. Internet n'est pas le Web

Le Web ne doit pas être confondu avec **Internet qui existait déjà avant l'apparition du Web**.

Internet est un **réseau informatique mondial qui interconnecte tous (ou presque) les réseaux locaux existant à travers le monde**.

Un peu comme le réseau routier, Internet est juste un ensemble de routes numériques par lesquelles on peut faire voyager toutes sortes de données à travers le monde entier.

Et tout comme cohabitent sur nos routes des vélos, des voitures, des camions, des motos, ... **différents services cohabitent également sur Internet** : le Web, les mails, le transfert de fichiers de données, les données de fonctionnement d'Internet, ...


#### 3. Le protocole HTTP

Avant le Web, des données circulait déjà sur internet, mais ce que Tim Berners-Lee a inventé, c'est un trafic de données d'un nouveau genre, basé sur des **mots cliquables** appelés **hyperliens**.

Ces hyperliens permettent de **naviguer d'une page d'informations à une autre**, grâce à un **protocole nommé HTTP** et inventé pour l'occasion.

En effet, si demain on introduit un nouveau type de véhicule sur nos routes, il faudra bien aussi définir des nouvelles règles de circulation qui leur soient adaptées.

Les **hyperliens sont écrits sous la forme d'URL** (Uniform Resource Locator) qui indiquent **où et comment trouver la page web visée**. Par exemple, l'URL suivante permet de charger la toute première du web, telle qu'elle a été publiée le 30 avril 1993 à la naissance du web :

http://info.cern.ch/hypertext/WWW/TheProject.html

#### 4. HTML et CSS

Mais pour pouvoir **mettre en forme et afficher ces pages d'informations** et les liens qu'elles contiennent, Tim Berners-Lee a aussi du développer un logiciel dédié. C'est ainsi qu'est né le premier **navigateur internet** ou navigateur web appelé **Mosaic**.


Ce navigateur web, tout comme ceux d'aujourd'hui (Chrome, Firefox, Safari, Opera, Brave, ...), a donc pour tâche de **lire les informations** que contient la page web **mais aussi des instructions** qui indiquent **comment elles doivent être affichées** (police, taille, couleur, justification, ...).

Pour indiquer l'ensemble des **informations et des instructions d'affichage qui leur sont associées**, on utilise conjointement les langages **HTML et CSS**.


La partie **HTML** contient les **informations** elles-mêmes (textes, images, vidéos, ...) **et la structure de la page** (paragraphes, titres, tableaux, listes à puces, ...) et c'est le **CSS** qui contient les **instructions de mise en forme** (taille, couleur et type de police, couleur de fond, marges, ...).

#### 5. Web statique et dynamique

Petit à petit, les fonctionnalités du web ont été développées afin de pouvoir afficher des gifs, puis des photos, puis de la vidéo, puis du contenu interactif et enfin des pages personnalisées.

On est ainsi passé progressivement du **web statique**, où le **contenu d'une page est le même pour tout le monde et tout le temps**, au **web dynamique** où **le contenu d'une page est généré à la volée de façon à s'adapter à chaque internaute et à l'instant**.

Mais pour en arriver là, il a fallu enrichir les outils disponibles pour le web. Par exemple le **Javascript** est un **langage de programmation classique** (boucles, conditions, variables, fonctions, ...) **intégré aux pages HTML** pour générer de l'interactivité (boutons, animations, ...). Ou encore le **PHP** est un langage qui permet de **générer des pages HTML personnalisées** à partir de données extérieures, comme des bases de données par exemple.


### II. Les URL

#### 1. Définition


Une **URL** ou **adresse web** n'est rien d'autre que le **chemin d'accès à une ressource accessible sur le Web**, le plus souvent un fichier.

C'est elle qui s'affiche dans la **barre d'adresse du navigateur web**.


**Remarque :** En réalité les URL permettent aussi d'accéder à des ressources qui ne sont pas sur le Web. Par exemple, une URL qui commence par "file://" pointe vers un fichier enregistré en local sur l'ordinateur de travail.

#### 2. Notion de chemin

Le **chemin** d'un fichier indique sa **position dans l'arborescence** (arbre à l'envers ici) du support de stockage.

Le *fichier6.html* est dans le *dossier4* qui est dans le *dossier2* qui est à la **racine** (notée slash "/"), c'est-à-dire tout en haut (ou tout en bas) du système de fichiers.

#### 3. Chemin relatif et chemin absolu

Pour indiquer le chemin d'un fichier, on peut le faire :
- **A partir de l'endroit où l'on se situe** actuellement, on parle alors de **chemin relatif**. Par exemple, si on est dans *dossier2*, le chemin relatif de *fichier6.html* est `dossier4/fichier6.html`.
- Ou bien **à partir de la racine**, on parle alors de **chemin absolu**. Le chemin absolu de *fichier6.html* est `/dossier2/dossier4/fichier6.html` et ce, quelque soit l'endroit où l'on se trouve.


Pour indiquer le **chemin relatif** vers un **fichier situé plus haut dans l'arborescence** (plus proche de la racine), on utilise ".." qui correpond au **dossier parent**, c'est-à-dire au dossier situé juste au-dessus de la position actuelle.

Ainsi, si on est dans *dossier2*, le chemin relatif de *fichier4.js* est `../dossier1/fichier4.js` et si on est dans *dossier3*, c'est `../../dossier1/fichier4.js`. Dans le premier cas, on remonte au dossier parent qui est la racine. Dans le deuxième, il faut remonter deux fois vers le dossier parent avant d'arriver à la racine qui contient *dossier1*.

**Remarque :** La syntaxe utilisée ici pour les chemins est celle des systèmes UNIX, Linux ou MacOS. Sous DOS (et donc sous Windows), le "\" (antislash) replace le "/".
