* **Faire afficher le code ASCII d'un caractère**

L'instruction Python pour faire afficher le code ASCII d'un caractère est la fonction `ord(char)` :

````python
print(ord("A"))
````

* **À l'inverse :**

La fonction `chr(n)` renvoie le caractère numéro `n` :

````python
print(chr(65))
````

* **Accès aux lettres d'une chaîne de caractères**

Pour accéder aux lettres d'une chaîne de caractères, on peut utiliser une simple boucle `for`.

````python
chaine="ceci est une chaîne de caractères"
for lettre in chaine:
  print(lettre, end=" ")
````
