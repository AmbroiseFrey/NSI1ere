## 2. Connecter des sous-réseaux

Dans la première partie, nous avons crée un sous-réseau de 3 machines appartenant toutes au sous-réseau 192.168.0.0 et avec les numéros de machines 1, 2 et 3 et reliées par un switch.

**Ajouter un deuxième sous-réseau d'adresse 192.168.1.0 comportant aussi 3 machines.** Vous êtes libre de choisir leurs numéros et leurs noms.

**Relier les 2 switchs et essayer de pinger une machine du deuxième sous-réseau à partir d'une machine du premier sous-réseau.** Cela ne fonctionne pas. :(

Pour relier deux sous-réseau entre eux, il faut **utiliser un routeur**. Cette [vidéo](https://youtu.be/xyK6ThdQeR0) montre comment faire.

**Remarques :**
- Il ne **suffit pas d'ajouter le routeur** pour que cela fonctionnne. Contrairement à un switch, un **routeur doit être configuré**.
- Le **routeur a deux interfaces réseaux et donc deux adresses IP** (il peut même en avoir plus), une dans chacun des sous-réseaux qu'il connecte ensemble. Ce n'est pas une obligation mais on a l'habitude de donner comme adresse IP au routeur la première (.1) ou la dernière (.254) IP disponible (.0 c'est le nom du sou-réseau et .255 est réservé pour le *broadcasting*).
- Le routeur jouant le rôle de passerelle entre les 2 sous-réseaux, son adresse IP (dans le bon sous-réseau) doit être **indiquée comme passerelle pour toutes les machines des 2 sous-réseaux**.
- Ne pas oublier de cocher *Routage automatique* pour que le routeur face son travail convenablement.

## 3. Installer des serveurs


### 3.1 Serveur DNS

Un **serveur DNS** (Domain Name System) permet d'**associer des noms de domaine à des IP**. Ainsi, il devient possible d'**envoyer des paquets IP à une machine en connaissant seulement son nom**, sans connaître son adresse IP.

On continue pour cela **à partir du réseau Filius précédent** qui comporte au moins deux sous-réseaux de 3 machines connectées par un routeur. On ajoute un *ordinateur* que l'on relie directement au **routeur** après avoir ajouté à ce dernier un **troisième connecteur réseau**. Au niveau IP, cet ordinateur se configure exactement comme si c'était un nouveau sous-réseau à lui tout seul.

Pour **plus de détails**, voir cette [vidéo](https://youtu.be/K3GGmiLwB6U).

Reste ensuite à **installer et configurer un serveur DNS** sur cet ordinateur en précisant quel nom de domaine (par exemple www.google.fr) est associé à quel IP (vous pouvez si vous le souhaitez, pour plus de réalisme, renommer une de vos machines "www.google.fr"). Il n'est **pas nécessaire d'associer un nom de domaine à chacune des machines** du réseau, une seule suffit ici. Par contre, **attention à ne pas oublier de démarrer le serveur DNS !**

Pour terminer, on **complète la configuration IP** d'une autre machine de notre réseau en lui précisant l'adresse IP du serveur DNS crée et on test le résultat en tapant `ping www.google.fr` (si on a utilisé ce nom de domaine là, sinon mettez le nom choisi) dans l'invite de commande de cette machine.

**Remarque :** Evidemment, cela n'a pas grand intérêt d'utiliser un serveur DNS pour un seul nom de domaine. Dans la réalité, **les serveurs DNS connaissent les adresses IP d'un très grand nombre de domaines**.

### 3.2 Serveur Web

Sur le même principe, **ajouter à votre réseau un serveur web**. Une fois de plus, vous pouvez vous inspirer de cette [vidéo](https://youtu.be/EZp_TLGVyv0).

Contrairement à la vidéo, vous allez avoir **en même temps le serveur Web et le serveur DNS**. Il serait donc de bon ton de modifier le serveur DNS ou d'y ajouter un **nom de domaine** (www.toto.fr par exemple) **qui pointe sur votre tout nouveau  serveur Web** (pas montré dans la vidéo).

Quand vous créez votre serveur Web, celui-ci dispose déjà d'une page web *index.html* (en allemand) qui lui permet de fonctionner. Mais n'hésitez pas à la personnaliser avec votre propre code HTML. Vous pouvez même **importer quelques petites images** avec la fonction *Importer* de l'explorateur de fichiers, mais n'en faites pas trop quand même. Là encore, **n'oubliez pas de démarrer le serveur Web**.

Une fois le serveur installé, vous n'avez plus qu'à **installer un navigateur web sur la machine de votre choix** et à **taper l'adresse IP ou le nom de domaine** (si vous avez bien renseigné l'IP du serveur DNS dans la configuration IP de cette machine) **de votre serveur Web** dans la barre d'adresse pour voir s'afficher la page index.html que vous avez créée.

Pour terminer, **utilisez la commande** *Afficher les échanges de données* (clic droit sur la machine avec le navigateur web, voir vidéo) afin de **visualiser les données échangées sur le réseau**. Chaque **ligne** qui s'affiche correspond à une **trame**, un **paquet** ou **datagramme** envoyé sur le réseau. Vous pouvez voir de quelle **couche** et de quel **protocole** il s'agit, ainsi que les IP de l'emetteur et du destinataire.

En **cliquant sur une ligne**, vous pouvez avoir plus de détails et retrouver les **en-têtes** ajoutées par chacune des couches concernées. Le plus intéressant est de **rechercher la requête** *HTTP GET* émise par le navigateur web ainsi que la réponse renvoyée par le serveur web. Vous pourrez notamment y retrouver le **contenu de la page HTML** renvoyée par le serveur.

Remarquez aussi les *segments* envoyés par TCP. Ceux dont la colonne *Commentaires* commence par *ACK* sont des **accusés de réception**.
