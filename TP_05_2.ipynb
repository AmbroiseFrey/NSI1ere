# <span style="color:red;"><center>TP - Créer une page web avec CSS </center></span>

## 1. Objectif du TP

On veut cette fois créer une **page web un peu plus complexe**, en utilisant le CSS.

## 2. Le TP de OpenClassRooms

Cette fois, nous allons directement utiliser l'excellent site d'[OpenClassRooms](https://openclassrooms.com/fr/). Ce site propose des formations de type **MOOC** dont certaines en **libre accès**. C'est notamment le cas de leurs formations sur [HTML5 et CSS](https://openclassrooms.com/fr/courses/1603881-apprenez-a-creer-votre-site-web-avec-html5-et-css3), [JavaScript](https://openclassrooms.com/fr/courses/6175841-apprenez-a-programmer-avec-javascript) ou encore [Java](https://openclassrooms.com/fr/courses/6173501-debutez-la-programmation-avec-java) et [Github](https://openclassrooms.com/fr/courses/5641721-utilisez-git-et-github-pour-vos-projets-de-developpement).

Pour y accéder, il suffit de se créer un compte gratuitement.

Nous n'allons **pas faire en classe la formation complète sur HTML et CSS** qui prend une vingtaine d'heures mais si vous avez le temps et si ça vous intéresse, n'hésitez pas. Nous allons juste faire le **TP qui permet de créer une belle page web**:

[TP : Créez un site pas à pas](https://openclassrooms.com/fr/courses/1603881-apprenez-a-creer-votre-site-web-avec-html5-et-css3/1606688-tp-creez-un-site-pas-a-pas)

**Sauf pour le dépot de votre travail bien sûr.** Celui-ci sera à déposer sur **Atrium** (seulement le fichier HTML) dans le **casier "Page sans CSS"**.
