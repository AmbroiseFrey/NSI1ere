## Introduction

Jusqu'à présent, nous avons créé des **pages web** essentiellement **passives** (non-interactives), qui se contentaient de **délivrer de l'information**. L'interaction avec l'utilisateur était limitée à cliquer ou pas sur des **liens hypertextes**.

Nous allons à présent voir comment on peut rendre une **page web interactive** en utilisant un langage de programmation comme le **JavaScript**.

**ATTENTION :** Il ne s'agit pas ici d'apprendre à programmer en JavaScript, comme nous l'avons fait pour Python, mais seulement de le voir en fonctionnement sur quelques exemples.

Ceci dit, JavaScript est un **langage impératif**, tout comme Python. On y retrouve donc tous les ingrédients que vous connaissez bien comme les variables, les fonctions, les boucles, les conditions, ... Pour ceux qui souhaitent en apprendre davantage sur le Javascript, vous pouvez aller sur ce [cours openclassrooms](https://openclassrooms.com/fr/courses/6175841-apprenez-a-programmer-avec-javascript?archived-source=2984401).

## B. Créer une page interactive avec JavaScript

Le langage JavaScript a été **créé en seulement dix jours** par *Brendan Eich* en 1995. Il travaillait alors pour la *Netscape Communcations Corporation*, à l'origine de **Netscape**, l'un des premiers navigateurs grand public.

Malgré son nom, JavaScript n'a **rien à voir avec le langage Java**, même si les deux langages peuvent être proches au niveau syntaxique (dans leur écriture).

Aujourd'hui, JavaScript est devenu **incontournable dans le développement web**.

### I. Un premier exemple trivial

#### 1. Le fichier HTML

Commencez par créer un fichier *index.html* contenant le code suivant :

    <!doctype html>
    <html lang="fr">
        <head>
            <meta charset="utf-8">
            <title>Le trio</title>
            <link rel="stylesheet" href="style.css">
        </head>
        <body>
            <h1>Le trio : HTML, CSS et JavaScript</h1>
            <p>Voici une page web qui ne fait pas grand-chose</p>
        </body>
        <script src="script.js"></script>
    </html>

En dehors de ce que nous avons déjà vu, la **seule nouveauté** est la ligne ``<script src="script.js"></script>``.

Comme vous vous en doutez, la balise ``<script>`` permet d'**insérer du code JavaScript**. Ici, on se contente d'insérer l'ensemble du code JavaScript contenu dans le fichier *script.js*, qui doit bien sûr se trouver dans le même dossier que *index.html*.

Vous pouvez **ouvrir le fichier HTML dans un navigateur**, mais comme nous n'avons pas encore crée le fichier JavaScript, cela ne **fera donc rien pour le moment** (à part un message d'erreur dans la console).

Vous avez peut être aussi remarqué que notre code HTML fait appel à un fichier CSS *style.css* qu'il va donc aussi nous falloir créer. Nous allons commencer par lui.

#### 2. Le fichier CSS

**Dans le même répertoire**, créez à présent le fichier *style.css* contenant le code (très restreint !) ci-dessous :

    h1{
        text-align: center;
    }

Vous le savez, cette règle CSS a simplement pour effet de centrer sur la page tous les titres de niveau 1. Vous pouvez le vérifier en ouvrant le fichier *index.html* avec un navigateur ou en rafraichissant la page si vous l'aviez déjà ouvert.

#### 3. Le fichier JavaScript

Pour finir, **toujours dans le même répertoire**, créez à présent le fichier *script.js* contenant le code (tout aussi restreint !) ci-dessous :

    alert("Le JavaScript fonctionne !")

En rafraichissant le fichier *index.html* dans le navigateur, vous constatez alors l'ouverture d'une *pop-up* indiquant *Le JavaScript fonctionne !*.

**Bravo, vous venez d'exécuter votre premier programme JavaScript !**

C'est évidemment l'effet de la fonction `alert()` en JavaScript qui permet de déclencher l'**ouverture d'une pop-up avec le texte de son choix fourni en argument**.

### II. Premier exemple d'interactivité

Dans notre premier exemple, la pop-up s'ouvre automatiquement à l'ouverture de la page HTML, **comment faire pour que ce soit une action de l'utilisateur qui en déclenche son ouverture ?**

#### 1. Création d'un bouton en HTML

Commencons par **modifier le code HTML** en y **insérant un bouton cliquable** grâce à la balise `<button>`. Pour cela il suffit d'insérer dans le *body* la ligne de code suivante :

	<button onclick="maFonction()">Cliquer ici</button>

Le texte *Cliquer ici* s'affichera sur le bouton et le fait de **cliquer sur ce bouton appelera une fonction JavaScript** appelée `maFonction()`. Comme à ce stade, cette fonction n'existe pas, cliquer sur le bouton ne fera rien.

**ATTENTION : A chaque fois que vous modifiez un des fichiers, il est évidemment nécessaire de le sauvegarder afin de pouvoir observer des changements dans le navigateur !**

#### 2. Création d'une fonction en JavaScript

Pour que le boutton fasse effet, il faut maintenant **créer une fonction Javascript** `maFonction()` et cela se fait bien sûr dans le fichier *script.js*.

**Remplacer le code précédent** par celui-ci :

    function maFonction() {
        alert("Le JavaScript fonctionne !")
    }

Ce code permet de **définir la fonction** souhaitée (`function` est équivalent à `def` en Python) et cette fonction se contente d'**ouvrir la même pop-up que précédemment**.

En rafraichissant le navigateur, vous ne devriez plus avoir de pop-up. Par contre, il suffit de cliquer sur le bouton *Cliquer dessus* pour la faire apparaître.

**Bravo ! Vous venez de créer votre première page HTML interactive !!**

### III. Modifier le style "à la volée"

Evidemment JavaScript ne sert pas qu'à ouvrir des pop-up. En fait on peut faire à peu près n'importe quoi en JavaScript. Voyons comment **modifier le style de notre page HTML** par exemple.

#### 1. Utilisation des identifiants

Pour que le JavaScript puisse modifier un élément de la page HTML, il faut qu'il **sache de quel élément il s'agit**. On peut utiliser pour cela les **identifiants** (`id = ...`).

Notre page HTML contient un seul **paragraphe** pour le moment. Attribuez lui l'identifiant *monPara* en utilisant la balise ouvrante `<p id="monPara">`.

#### 2. Modifier le style en JavaScript

En JavaScript, l'objet *document* correspond à la **page HTML** courante.

La méthode *.querySelector(id)*, appliquée à *document*, permet de **récupérer l'élément** dont l'identifiant est *id*.

La méthode *.style*, appliquée à un élément, **récupère le style** de cet élément.

Et enfin, la méthode *.color="couleur"*, appliquée à un style, permet de **modifier la couleur** définie dans ce style.

**Remplacer à nouveau** le contenu de *script.js* par le code suivant :

    function maFonction() {
        document.querySelector("#monPara").style.color="red";
    }

Cette nouvelle fonction JavaScript doit donc **modifier la couleur d'écriture de l'élément dont l'id est *monPara* pour le mettre en rouge**.

Après avoir enregistré les fichiers modifiés, rafraichir la page HTML et **vérifier qu'en cliquant sur le bouton, le paragraphe passe bien en rouge**.

Evidemment, on peut, de la même façon, **modifier n'importe quelle autre règle de style de n'importe quel élément de la page**.

Mais comme on l'a vu, on préfère souvent en CSS créer des **classes** qui regroupent un **ensemble de règles de style** (couleur, police, taille, ...). Par conséquent, il serait préférable de **changer la classe de l'élément** pour modifier toutes ses règles de style en une seule fois.

#### 3. Utilisation des classes

Commencons par créer 2 classes *.rouge* et *.vert* dans notre fichier CSS.

Pour cela, ajoutez dans *style.css* le code suivant :

    .rouge {
        color:red;
        font-size:20px;
    }
    .vert {
        color:green;
        font-size:30px;
    }

#### 4. Modifier la classe en JavaScript

Remplacer le code JavaScript de *script.js* par le code suivant :

    function foncRouge() {
        document.querySelector("#monPara").classList.remove("vert");
        document.querySelector("#monPara").classList.add("rouge");
    }
    function foncVert() {
        document.querySelector("#monPara").classList.remove("rouge");
        document.querySelector("#monPara").classList.add("vert");
    }

Sur le **même principe que précédemment**, ces 2 fonctions *foncRouge* et *foncVert* permettent d'**enlever et d'ajouter l'association des classes *rouge* et *vert*** avec l'élément d'identifiant *monPara*.

A vous de **modifier votre fichier HTML** en **créant 2 boutons marqués *Rouge* et *Vert*** qui auront pour effet d'**appliquer respectivement la classe *Rouge* et la classe *Vert* à notre paragraphe** (on en a toujours un seul) lorsque l'on clique dessus.

**Remarque :** Vous pouvez si vous le souhaitez **garder le bouton déjà existant** mais si vous avez effacé la fonction *maFonction* du fichier JavaScript, il sera sans effet. Par contre, si  vous le gardez ainsi que la fonction *maFonction* précédente, vous devriez observer un **conflit intéressant** entre les effets des 3 boutons. Pour expliquer ce comportement, il faut se rappeler ce que nous avions dit sur les **priorités en matière de régles CSS**.

### IV. Modifier le contenu d'un balise HTML

Non seulement JavaScript peut **modifier le style d'un élément HTML** mais il peut **aussi modifier son contenu**.

Notre cobaye étant toujours le paragraphe *monPara*, voyons comment modifier le texte de ce paragraphe, toujours en cliquant sur un bouton.

#### 1. Modification du fichier HTML

**Ajouter** le bouton suivant à votre code HTML :

		<button onclick="modifMessage()">Cliquez ici</button>

#### 2. Nouvelle fonction JavaScript

Ajouter la fonction suivante dans le fichier JavaScript :

    function modifMessage() {
        document.querySelector("#monPara").innerHTML = "Bravo, vous avez cliqué sur le bouton !"
    }

La méthode *.innerHTML* permet d'**accéder au contenu de l'élément HTML auquel elle s'applique** pour le remplacer par autre chose, ici un autre texte.

**Enregistrer les fichiers** HTML et JavaScript et **rafraichir** la page dans le navigateur pour **tester** ce nouveau bouton.

Pour l'instant, nous nous sommes contentés d'utiliser **l'événement** *Click sur un bouton* pour déclencher l'**exécution des fonctions JavaScript**. Mais il existe bien d'autres possibilités, comme le **survol de la souris** par exemple.
