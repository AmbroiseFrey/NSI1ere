## III. Les types simples
En informatique on manipule des **données**, elles peuvent être de **types simples** (nombres, mots, booléens(Vrai ou Faux)) ou comme nous le verrons plus tard de **types construits** (listes, tuples, dictionnaires) et enfin **structurées** (tableaux, base de données).  
- **Les nombres** :
    • Les **entiers** (positifs ou négatifs) de type `int`, pour integer.
    • Les **nombres à virgule** de type `float`, pour nombre à virgule flottante.
*_Remarque_ : Du fait de la limitation en mémoire, en informatique tous les réels sont des décimaux, il ne peut pas y avoir une infinité de chiffres après la virgule.*
- **Les chaînes de caractères** : Le type d’une chaîne de caractère est `str`, pour String.
- Les **booléens** : Le type d’un booléen est `bool`, pour boolean. Un booléen ne peut avoir que deux valeurs : `True` ou `False`.


**IMPORTANT :** En Python, le typage d’une variable est *implicite* : c’est à dire que le type d’une variable n'a pas être défini dès le départ et s'adapte à ce que la variable contient.


La fonction `type` permet de connaître le type d'une variable existante.

## V. Les instructions conditionnelles
Une instruction conditionnelle est une instruction que ne s'éxecute que si une certaine condition est réalisée. Cette condition s'écrit en général sous la forme d'un **test de comparaison**. En général, l'instruction conditionnelle n'est exécutée que si le résultat de ce test est vrai.
### V.1 Les opérateurs de comparaison
 Soient x et y deux **variables numériques**, alors on peut tester :
 - leur **égalité** avec l'opérateur `x == y`
 - leur **non-égalité** avec l'opérateur `x != y`
 - le **supérieur ou égal** avec l'opérateur `x >= y`
 - le **plus petit ou égal** avec l'opérateur `x <= y`
 - la **stricte supériorité** avec l'opérateur `x > y`
 - la **stricte infériorité** avec l'opérateur `x < y`
 
Dans tous les cas, ces tests ne peuvent valoir que True ou False ; ce sont des **booléens**.
 
*Remarques :*
- Les tests d'égalité et de non égalité marchent aussi avec des chaîne de caractères ou des booléens.
- Ils sont par contre à éviter avec des nombres flottants (nombres à virgules) car les erreurs d'arrondis peuvent rendrent différents des nombres qui devraient être égaux (voir IV.1).

### V.2 La structure conditionnelle
En python, une structure conditionnelle s'écrit en toute généralité de la façon suivante :
> ``` python
if test == True :
    instruction1
else :
    instruction2
```

Cela signifie que `instruction1` sera exécutée si le résultat du test est vrai et que sinon, c'est l'`instruction2` qui sera exécutée.
Mais comme la véracité du test est la condition par défaut pour exécuter la première instruction, le code précédent est complétement équivalent au code suivant :
> ``` python
if test :
    instruction1
else :
    instruction2
```

**<span style="color:red;">IMPORTANT :</span>** Remarquez la présence du caractère ":" à la fin des lignes commençant par `if` et `else`, ainsi que le décalage, appelé **indentation**, des lignes qui suivent ce caractère. Cela peut sembler anodin mais sans cela, le programme ne fonctionnera pas, ou pire, il fonctionnera mal !

*Remarques :*
- La présence du `else` n'est pas obligatoire.
- On peut tout à fait exécuter l'`instruction1` si le résultat du test est faux en mettant comme condition `test == False` ou `test != True` (la première expression est conseillée).
- On peut aussi combiner plusieurs tests simultanément avec les opérateurs logiques `and`, `or`et `not` ("et", "ou" et "non"). Par exemple, l'expression `(test1 and test2) == true` n'est vraie que si test1 et test2 sont vrais simultanément.
