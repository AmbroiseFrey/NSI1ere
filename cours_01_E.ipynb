## E. Représentation binaire des caractères

### I. Les caractères alphanumériques

Les caractères alphanumériques sont l'ensemble des symboles qui composent un texte. Ils se composent :
- des lettres majuscules et minuscules, accentuées ou non,
- des symboles de ponctuation ou autre ($ & , ; : £ # ...),
- des chiffres.

**Attention** à ne pas confondre le caractère "1" et l'entier 1. Comme ce ne sont pas les mêmes types, l'opérateur $+$, par exemple, n'aura pas du tout le même effet : concaténation pour les caractères et addition pour les entiers. 

### II. Le codage ASCII

#### 1. La table ASCII

**Exemples :**
- Le codage de la lettre "f" est 0110 0110 = 102. Le code ASCII de f est donc 102.
- Vérifier que le codage de "a" est 97 et celui de "z" est 122.

**Remarques :**
- Ce tableau ne comporte pas de caractères accentués puisqu'il a été conçu pour la langue anglaise. C'est de là que vient par exemple que ces caractères sont interdits dans les adresses mail.
- Les deux premières lignes correspondent à des caractères de contrôle utilisés par les premiers ordinateurs. Peu sont encore utilisés mais on y reconnait quand même $ESC$ qui correspond à la touche $Echapp$ de nos claviers ou $BS$ pour $Backspace$ ou encore $DEL$ pour $Suppr$.
- La table ASCII ne comporte que 128 caractères car, à l'origine, les caractères étaient codés sur 7 bits. Mais la généralisation de l'utilisation de l'octet a conduit à coder les caractères sur 8 bits. C'est pourquoi tous les codes binaires commencent par 0.


#### 4. Groupes de caractères

Les caractères de 0 à 31 et le caractère 127 sont les seuls non affichables sur un écran.
Les lettes se suivent dans l'ordre alphabétique (65 à 90 pour les majuscules et 97 à 122 pour les minuscules). Il faut donc ajouter ou retrancher 32 pour passer de minsucule à majuscule et inversement.




Il est très important de comprendre que du fait cette **limitation de la quantité de mémoire allouée au stockage des nombres**, un ordinateur ne fait, le plus souvent, des **calculs exacts qu'avec les nombres entiers**. Ainsi, pour un ordinateur, $\sqrt{2}^2 \ne 2$ :

```python
# La preuve
from math import *
print(sqrt(2)**2)
print (sqrt(2)**2 == 2)
```

### II. Convertir un nombre réel en binaire

#### 2.1 Ecriture d'un nombre décimal en base 10 (programme de 5ème)
Quand on écrit le nombre 23,375 en base 10, cela signifie que ce nombre s'obtient en faisant la somme de 2 dizaines, plus 3 unités, plus 3 dixièmes, plus 7 centièmes, plus 5 millièmes.

Soit mathématiquement :

$$ 23,375 = 2 \times 10^1 + 3 \times 10^0 + 3 \times 10^{-1} + 7 \times 10^{-2} + 5 \times 10^{-3}$$


#### 2.2 Ecriture d'un nombre dyadique en binaire
On peut très bien faire exactement la même chose en binaire. On décompose la **partie entière** en une somme de **puissances positives de 2** (c'est l'écriture binaire classique d'un nombre entier) et la **partie décimale** en une somme de **puissances négatives de 2** :

- $23 =  1 \times 2^4 + 0 \times 2^3 + 1 \times 2^2 + 1 \times 2^1 + 1 \times 2^0 = 16 + 4 + 2 + 1 = 23$

- $0,375 = 0 \times 2^{-1} + 1 \times 2^{-2} + 1 \times 2^{-3} = 0,25 + 0,125 = 0,375$

- On a donc : $23,375_{10} = 10111,011_2$

En fait, cela fonctionne dans cet exemple car 23,375 est un **nombre dyadique**, c'est-à-dire qu'il peut s'écrire sour la forme $\frac{N}{2^{\, p}}$ où $N \in \mathbb{Z}$ et $ p \in \mathbb{N}$. Ce sont les **seuls nombres dont l'écriture en binaire est finie**.

#### 2.3 Méthode de conversion
Comme pour le codage des entiers, il existe une méthode astucieuse pour **trouver facilement le codage binaire de la partie décimale** : il suffit de **multiplier la partie décimale par 2 jusqu'à obtenir 1** et de **noter la partie entière du résultat obtenu à chaque étape**.

Exemple pour 0,375 :
- $0,375 \times 2 =$ <span style="color:red;">0</span>,75
- $0,750 \times 2 =$ <span style="color:red;">1</span>,5
- $0,5 \times 2 =$ <span style="color:red;">1</span>
- On retrouve bien : $0,375_2 =0,011_2$


**Application**

Convertir en écriture binaire les nombres suivants : 5.25, 0.5, 12.75 et 128.375.

#### 2.4 Cas des nombres non dyadiques

Si on essaye de coder en binaire, selon la méthode précédente le nombre 0,1 on obtient :
- $0,1 \times 2 =$ <span style="color:red;">0</span>,2
- $0,2 \times 2 =$ <span style="color:red;">0</span>,4
- $0,4 \times 2 =$ <span style="color:red;">0</span>,8
- $0,8 \times 2 =$ <span style="color:red;">1</span>,6
- $0,6 \times 2 =$ <span style="color:red;">1</span>,2
- $0,2 \times 2 =$ <span style="color:red;">0</span>,4
- $0,4 \times 2 =$ <span style="color:red;">0</span>,8
- $0,8 \times 2 =$ <span style="color:red;">1</span>,6
- ...
- D'où : $0,1_{10} =0,0 0011\, 0011 \, 0011 ..._2$

En binaire un **nombre non dyadique** s'écrit avec un **nombre infini de bits**, de même qu'en base 10 un **nombre non décimal** s'écrit avec un **nombre infini de chiffres**.

**Remarque**

On retrouve pour les **nombres non dyadiques le problème de calcul approximatif** indiqué plus haut. Il est impossible de faire des calculs exacts en informatique avec des nombres non dyadiques puisqu'on ne manipule que des valeurs approchées de ces nombres.

```python
# Illustration
print(0.1)
print(0.1+0.1)
print(0.1+0.1+0.1)
```

**Application**

Convertir en écriture binaire les nombres suivants : 17.2, 35.31 et 0.05.

### III. Notation en virgule flottante

#### 3.1 Ecriture scientifique

Dans ce qu'on a vu au II., il manque une façon de coder **la position de la virgule**. Vous savez que la **notation scientifique** permet d'uniformiser l'écriture des nombres décimaux sous la forme $\pm a \times 10^n$ où $a$ est un nombre décimal entre 1 et 10 (non inclus) appelé **mantisse** et $n$ un entier relatif appelé **exposant**.

De même, on peut toujours écrire un nombre décimal binaire sous la forme $\pm b \times 2^n$ où $b$ est un nombre binaire de la forme $1,...$ et $n$ un entier relatif.

#### 3.2 Le typle `float 32`
Pour coder un nombre décimal $nombre$, on a alors seulement besoin de coder les nombres $signe$, $mantisse$ et $exposant$ défini par :
$$ nombre=(-1)^{signe}\times 1,mantisse\times 2^{n+127}$$
Il ne reste plus qu'à définir le nombre de bits alloués pour chacun de ces nombres. Par exemple, pour le type `float`, le plus courant, codé sur 32 bits au total, on a :
- le 1er bit pour le $signe$ (1 si <0 et 0 si >0)
- les 8 bits suivants pour l'exposant ($n+127$)
- les 23 bits restant pour la mantisse



#### 3.3 Remarques
- En codant $exposant=n + 127$ au lieu de simplement coder $n$, on évite de devoir coder des exposants négatifs.
- Comme la mantisse a forcément le chiffre 1 avant la virgule, cela ne sert à rien de la coder.
- Si le nombre est dyadique, de sorte que la mantisse comporte moins de 23 chiffres, on complète avec des 0.
- En base 2, comme en base 10, à chaque fois qu'on décale la virgule d'un cran vers la gauche, il faut multiplier le nombre par 2 pour conserver l'égalité.

#### 3.4 Exemple
On veut coder en `float 32 bits` le nombre décimal 211.3.
- Ce nombre est positif donc $signe = 0$.
- En écriture binaire : $211.3_{10} \approx 11010011.0100110011001101_2 \approx 1.10100110100110011001101_2\times 2^7$ donc la mantisse est $10100110100110011001101$.
- L'exposant est $7$ donc $7+127 = 134_{10} = 10000110_2$.
- On obtient donc le code binaire suivant sur 32 bits : $0\: 10000110 \: 10100110100110011001101$

**ATTENTION**

En réalité, ce nombre correspond à la valeur approchée $211.3000030517578125$.

**Application**

Donner le codage en binaire sur 32 bits de 35,31.

### IV. Flottants simple et double précision

####  4.1 Tester l'égalité de 2 nombres flottants

Comme on l'a vu, **la représentation binaire des nombres réels n'est qu'approximative**, il faut donc se montrer très méfiant si on a besoin de tester l'égalité de 2 nombres réels. De manière générale, il ne faut pas le faire sous peine de mauvaises surprises. Par contre, on peut **tester si la différence entre 2 nombres décimaux est plus petite qu'un certain seuil** (marge d'erreur) judicieusement choisi.


```python
# Exemple
print (0.1+0.2 == 0.3)
print (0.1+0.2-0.3 < 10**(-10))
```

#### 4.2 Les doubles précisions

Si l'on souhaite **plus de précision** sur la représentation binaire des nombres décimaux, il faut simplement utiliser plus de bits. Ainsi les **flottants double précision** (on parle souvent de *double*) sont codés sur **64 bits**.
