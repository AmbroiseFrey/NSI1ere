## <span style="color:blue;"> Programme de l'enseignement *Numérique et Sciences Informatiques* </span>

*Définition* : En **informatique**, on conçoit des **algorithmes** (suite d'opérations) qui opèrent sur des **données** (nombres, texte, ...) mais pour faire exécuter ces algorithmes par une **machine** (ordinateur), il faut d'abord les traduire dans un **langage** (python) qui soit compréhensible par la machine.


L'enseignement de NSI s'appuie sur ces **quatre** concepts clé de l'informatique :

* Les **données**, qui représentent sous une forme numérique unifiée des informations très diverses : textes, images, sons, mesures physiques, sommes d'argent, etc.

* Les **algorithmes**, qui spécifient de façon abstraite et précise des traitements à effectuer sur les données à partir d'opérations élémentaires.

* Les **langages**, qui permettent de traduire les algorithmes abstraits en programmes textuels ou graphiques de façon qu'ils soient exécutables par les machines.

* Les **machines**, et leurs systèmes d'exploitation, qui permettent d'exécuter des programmes en enchaînant un grand nombre d'instructions simples, assurent la persistance des données par leur stockage et de gérer les communications. On y inclut les objets connectés et les réseaux.

Le programme officiel comporte **7 thèmatiques** principales :
1. Représentation des données: types et valeurs de base
2. Représentation des données: types construits
3. Traitement de données en tables
4. Interactions entre l’homme et la machine sur le Web
5. Architectures matérielles et systèmes d’exploitation
6. Langages et programmation
7. Algorithmique

Pour le programme officiel complet, voir [ici](https://cache.media.eduscol.education.fr/file/SP1-MEN-22-1-2019/26/8/spe633_annexe_1063268.pdf).

## <span style="color:blue;">Les bonnes pratiques du programmeur</span>

### La sauvegarde

Vous devez impérativement **sauvegarder régulièrement vos travaux**. Le mieux est de le faire en ligne, que ce soit dans github, dans un cloud ou dans votre espace personnel sur Atrium.
Mais au cas où, il vaut mieux avoir aussi un support amovible (clé USB, disque dur, etc). Il est plus sûr d'avoir plusieurs sauvegardes différentes.

### Les noms de fichiers

Il est important de donner un **nom qui fasse sens à ses fichiers** pour savoir facilement ce qu'on peut y trouver à l'intérieur. Pour éviter les problème de compatibilité, on évite :
* Les lettres accentuées et autres caractères spéciaux.
* Les espaces que l'on remplace par un "_"(*mon_fichier* plutôt que *mon fichier*)

### S'organiser !

Il est impératif d'organiser vos productions avec une **arborescence claire**. Par exemple :
* NSI
    * Cours
    * TP
        * TP du chapitre 1
        * ...
    * Mes scripts
    * Exercices
        * Exos du chapitre 1
            * exo_1
            * etc.
            * ...
    * etc.
