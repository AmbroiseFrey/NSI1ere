## Introduction

Par définition, une **page web est écrite en HTML** (Hyper Text Markup Langage) et cela pourrait suffire, c'était d'ailleurs le cas au début du Web. Mais aujourd'hui, les **bonnes pratiques du développement web** demandent l'utilisation conjointe de **CSS** (Cascading Style Sheets).

Au maximum, il faut donc respecter la répartition suivante des rôles :
- Le fichier *.html* **contient l'information** (texte, images, liens, ...).
- Le fichier *.css* **contient les instructions de mise en page** (police, couleurs, marges, ...).

Cette distinction n'est pas toujours évidente car il est tout à fait possible de faire de la mise en page sans utiliser CSS, juste avec le HTML.

## A. Créer une page web statique

Une page web est dite **statique** si elle se contente d'afficher **la même information pour tous les utilisateurs**. De plus, on se restreindra pour le moment à des pages **sans interaction possible avec l'utilisateur** (en dehors des liens), c'est-à-dire **non-interactives**.

### I. Le HTML

#### 1. Un langage de balisage

Le HTML est un **langage informatique de description**, dit **de balisage**. Cela signifie que contrairement aux **langages de programmation** comme *Python* par exemple, il ne comporte aucune instruction pour faire des boucles ou des tests conditionnels, il ne manipule ni variable, ni fonction ...

NB :
- Pour le web, on utilse souvent le *JavaScript* comme langage de programmation.
- Le langage *Markdown* utilisé pour ce Notebook est aussi un langage de balisage mais on peut aussi très bien y écrire en HTML.

En HTML, on se contente de **structurer l'information avec des balises** qui indiquent **où commencent et se terminent** chaque élément : les paragraphes, les titres, les passages importants, les listes, les tableaux, ...

Pour cela, on utilise des **balises ouvrantes** de la forme ``<nom_balise>`` et des **balises fermantes** de la forme ``</nom_balise>``.

La première règle de base est donc : **Toute balise ouverte doit être fermée.**

NB : Comme toute règle, elle a ses exceptions ! ;)

#### 2. Le rôle du navigateur internet

Un **navigateur internet** (Firefox, Chrome, Opera, Safari, ...) est un **logiciel** capable de **lire** le texte et les instructions écrites dans un **fichier HTML et son fichier CSS associé**, afin de les interpréter et **d'afficher au mieux l'information contenue avec sa mise en page**.

Il est toujours possible de visualiser directement le **code source** d'une page web affichée par le navigateur. Sur PC, on peut en général y accéder avec la touche *F12* ou avec un clic-droit puis "Afficher le code source de la page".

De plus, si les fichiers sont hébergés sur le net, le navigateur gère la communication avec le serveur web (requêtes http).

#### 3. Premiers pas

Pour faire nos premiers pas en HTML, nous allons utiliser un **éditeur en ligne** qui a l'avantage de présenter dans la même fenêtre le code HTML, le code CSS (et aussi le JavaScript si besoin) et l'affichage que ferait un navigateur de ce code.

Le site choisi est https://jsfiddle.net/ car il ne demande aucune inscription.

##### 3.1 Hello World

Vous le savez, est informaticien digne de ce nom doit toujours commencer par faire afficher *Hello World* lorsqu'il commence un nouveau langage de programmation, sous peine d'être maudit sur 7 générations.

Sur le site https://jsfiddle.net/, dans la fenêtre *HTML*, vous allez donc commencer par copier-coller le code HTML ci-dessous. Puis cliquez sur le bouton *Run* en haut à gauche pour voir le résultat dans la fenêtre *Result* (c'est quand même bien fait, non ? ;) )

    <h1>Hello World! Ceci est un titre</h1>
    <p>Ceci est un <strong>paragraphe</strong>. Avez-vous bien compris ?</p>

**Remarques :**
- La règle de base sur les balises ouvrantes et fermantes est bien respectée.
- Les balises peuvent être **imbriquées** (on ouvre et on ferme une balise à l'intérieur d'une autre) mais **pas chevauchées** (si plusieurs balises sont ouvertes, il faut toujours commencer par fermer la dernière ouverte).
- La balise ``<strong>`` permet de mettre en évidence un passage important du texte. Le plus souvent, cette mise en évidence se traduira par une écriture en **gras**. Mais si ce n'est pas possible (tout le texte est déjà affiché en gras par exemple) alors cela pourra se traduire par un soulignement. Il vaut donc mieux utiliser la balise ``<strong>`` plutôt que la balise ``<b>`` (pour bold) qui ne fait que mettre en gras et qui n'est donc qu'un élément de style plutôt réservé au CSS.

##### 3.2 Les bonnes pratiques

**Beaucoup de choses sont autorisées** (ou pas interdites) par le code HTML (pas de message d'erreur quand vous chargez la page dans un navigateur) **mais très fortement déconseillées dans la pratique**.

Par exemple, la balise ``<h1>`` sert à écrire un titre (header) de niveau 1 qui, par défaut, apparaitra en plus gros qu'un titre de niveau 2 ouvert avec ``<h2>`` puisqu'il est plus important.  
Cependant, rien n'interdit d'inverser le rôle de ``<h1>`` et ``<h2>`` en modifiant la taille des polices (avec le CSS) ou encore d'écrire des paragraphes entier en *h1*.

**Mais il ne faut pas le faire**. Il faut respecter la **sémantique des balises**, cela fait partie des *bonnes pratiques* en HTML.

##### 3.3 Les attributs

La plupart des balises peuvent être ouvertes en précisant certains **attributs**, ce sont des **précisions sur le comportement de la balise** dont le rôle est très variable d'une balise à une autre.

Mais certains attributs, dits **globaux**, peuvent être utilisés avec **toutes les balises HTML**, c'est le cas de l'**identifiant** (*id*) et de la **classe** (*class*).

Ces attributs sont **utilisés conjointement avec le CSS pour mettre en forme** le contenu des balises.

Copier-coller le code HTML ci-dessous, à la place du précédent, sur le site *JSFiddle.net* et cliquez sur le bouton *Run*.

    <h1>Ceci est un titre</h1>
    <h2 class="titre_1">Ceci est un sous titre</h2>
    <p id="para_1">Ceci est un <strong>paragraphe</strong>.</p>
    <p>Avez-vous bien compris ?</p>

Pour l'instant, les attributs *class* et *id* n'ont aucune influence sur le rendu de la page. Pour changer cela, il faut passer par le CSS.

### II. Le CSS

#### 1. Un exemple

Copier-coller le code ci-dessous, mais cette fois **dans la fenêtre CSS** de *JSFiddle.net*.

    h1 {
        text-align: center;
        background-color: red;
    }
    h2{
        font-family: Verdana;
        font-style: italic;
        color: green;
    }
    #para_1{
        font-style: italic;
        color: green;
    }
    .titre_1{
        color:red
    }

Cliquer sur *Run* et observer le résultat.

#### 2. Structure d'un fichier CSS

Un fichier CSS **liste des sélecteurs** (*h1, h2, #para_1, .titre1*) qui **ciblent certaines balises HTML** et pour chacun desquels on indique des **règles de mise en forme** ou **règles de style** (alignement du texte, couleur de fond, nom de la police, style d'écriture, couleur d'écriture, ...).

On distingue dans notre exemple 3 sortes de sélecteur, **du plus général au plus précis** :
- Les **sélecteurs d'éléments** (par ex. *h1*) qui s'appliquent à toutes les balises d'un type donné (ici, toutes les balises de type *h1*).
- Les **sélecteurs de classe** (par ex. *.titre1*) qui s'appliquent à toutes les balises d'une classe spécifiques (ici, toutes celles qui comportent l'attribut *class="titre_1"*).
- Les **sélecteurs d'identifiant** (par exe. *#para_1*) qui s'appliquent à la seule et unique balise qui a cet identifiant (ici, celle qui a l'attribut *id="para_1"*).

#### 3. Les conflits en CSS

Avec ce système, il arrive fréquemment qu'une seule et même balise soit concernée par plusieurs règles de style qui rentrent en **conflit**.

Dans notre exemple, la balise ``<h2 class="titre_1">`` doit à la fois appliquer les règles qui s'appliquent à toutes les balises ``<h2>`` et celles qui s'appliquent à celles de la classe *titre_1*. Ce qui fait que le texte devrait apparaître à la fois vert et rouge ! Or on constate que le texte apparaît rouge, ce qui signifie que c'est la règle de classe qui est ici prioritaire.

Ceci nous amène à une règle de base de CSS : **En cas de conflit, c'est la règle la plus précise (la moins générale) qui s'applique.**

En terme de **priorité**, on a donc : **identifiant > classe > élément**

#### 4. Quel sélecteur utiliser en CSS ?

Les attributs *id* et *class* permettent tous les deux d'appliquer (ou de cibler) des règles de mise en forme à une balise. La différence fondamentale est que l'**on peut attribuer la même classe à plusieurs balises alors qu'un identifiant doit rester unique** (attaché à une seule balise).

En conséquence, on utilise les **classes pour définir des mises en forme qui se répètent dans la page** alors que les **identifiants servent à mettre en forme un élément qui n'apparaît qu'une seule fois** dans la page.

Les règles CSS définies pour un **sélecteur d'élément** sont les plus générales et donc à réserver à des **éléments de style qui ne varient pas de toute la page**, comme la police de caractère utilisée par exemple.

### III. Créer une page HTML avec CSS

#### 1. Méthodologie

Nous allons à présent abandonner JSFiddle qui permete seulement de tester des petits morceaux de code, pour nous atteler à la création d'une véritable page web.

Cela passe par la création d'un fichier HTML, par exemple *index.html*, et d'un fichier CSS, par exemple *style.css*. Ces 2 fichiers sont des fichiers texte que l'on créera et éditera avec l'éditeur *Notepad++* par exemple.

Pour voir ce que donne notre page web, il suffit de l'ouvrir avec un navigateur web (doucle clic sur le *index.html*). Ensuite, à chaque fois qu'on veut visualiser les modifications apportées sur nos fichiers HTML et CSS, il suffit de les sauvegarder puis de rafraichir la page sur le navigateur (touche F5).

#### 2. Première  page web

Dans *index.html*, copier-coller le code suivant puis le visualiser dans un navigateur.

    <!doctype html>
    <html lang="fr">
        <head>
            <meta charset="utf-8">
            <title>Voici mon site</title>
        </head>
        <body>
            <h1>Hello World! Ceci est un titre</h1>
            <p>Ceci est un <strong>paragraphe</strong>. Avez-vous bien compris ?</p>
        </body>
    </html>
    
**Remarque :** L'identation des blocks n'est pas obligatoire mais conseillée pour plus de lisibilité du code.

#### 3. Structure de base

L'exemple précédent comporte le minimum de toute page web qui se respecte :
- L'instruction ``<!doctype html>`` (qui n'est pas une balise) et indique seulement au navigateur que le fichier est en HTML 5.
- Les balises ``<html>`` et ``</html>`` sont obligatoires et encadrent l'ensemble du code HTML.
- Les balises ``<head>`` et ``</head>`` délimitent l'en-tête (*header*) qui contient quelques informations essentielles pour la page :
    - En particulier, la balise ``<meta charset="utf-8">`` précise l'encodage à utiliser pour les caractères et permet d'afficher correctement les caractères accentués. (Tiens, une balise qui ne se ferme pas !)
    - Quant aux balises ``<title>``et ``</title>`` elles permettent de définir le titre de la page qui s'affiche dans les onglets du navigateur.
- Enfin, les balises ``<body>`` et ``</body>`` délimitent le corps de la page, c'est-à-dire l'ensemble des informations que l'on veut afficher.

#### 4. Premier fichier CSS

Dans un autre onglet de *Notepad++*, copier-coller le code suivant et l'enregistrer sous *style.css* dans le même répertoire que le fichier *index.html*.

    h1 {
        text-align: center;
        background-color: red;
    }
    p {
        font-family: Verdana;
        font-style: italic;
        color: green;
    }

#### 5. Lier les fichiers HTML et CSS

Pour l'instant, le fichier CSS n'est pas utilisé car nous ne l'avons pas lier à notre fichier HTML. Pour cela, il faut **indiquer dans l'en-tête du fichier HTML le nom et le chemin du fichier CSS à utiliser**.

Ajouter la ligne suivante dans le *header*, enregister et rafraichir la page.

    <link rel="stylesheet" href="style.css">
    
NB :
- Comme les 2 fichiers sont dans le **même répertoire**, le nom du fichier CSS suffit. S'il était dans un sous-répertoire *CSS*, il faudrait écrire `` href="CSS/style.css"``.
- Que ce soit pour le **nom d'un fichier** CSS, d'une image ou d'un lien, les **caractères accentués**, les **espaces** et autres **caractères spéciaux** sont à **bannir** car leur traitement dépend trop de la machine sur laquelle le code sera exécuté.

### IV. Quelques balises HTML de base

#### 1. Insérer des liens hypertexte

Pour créer des **liens hypertexte**, on utilise la balise ``<a>`` et son **attribut** ``href`` qui sert à indiquer **vers quoi pointe le lien**. C'est tout le **texte contenu entre les balises ouvrante et fermante qui constitue le lien** (qui sera souligné en bleu sur la page).

Exemple de **lien vers un autre fichier HTML dans le même répertoire** :

    <a href="mon_autre_page.html">Cliquez ici pour vous rendre sur mon autre page</a>

Exemple de **lien vers une page internet** :

    <a href="http://www.google.fr">Cliquez ici pour vous rendre sur google.fr</a>
    

#### 2. Insérer des images

Pour **insérer une image**, on utilise la balise ``<img/>`` et ses **attributs** ``src`` qui indique où trouver l'image (nom et chemin si nécessaire) et ``alt`` qui est une description succinte de l'image. Cette balise est **à la fois ouvrante et fermante** (encore une exception) puisqu'elle sert à insérer une image à un endroit précis, elle n'est pas "*étendue*".

L'attribut ``alt`` est important car il permet de décrire l'image pour un non-voyant qui surfe sur le net à l'aide d'un navigateur muni de synthèse vocale ou il s'affiche à la place de l'image si celle-ci n'est pas trouvée.

Exemple d'**insertion d'une image** :

    <img src="mon_image.jpg" alt="avion"/>

Pour indiquer la taille d'affichage de l'image, on utilise les attributs ``width`` et ``height`` qui précise sa taille d'affichage en pixels.

Exemple :

    <img src="mon_image.jpg" width="233" height="217" alt="avion"/>

#### 3. Structurer la page

Les balises ``<div>`` et ``<span>`` sont un peu les balises à tout faire en HTML. Elles sont **génériques**, car elles ne servent à rien si ce n'est à **délimiter des morceaux de HTML** auxquels on voudrait par exemple appliquer un même style CSS. Idéalement, on ne **devrait pas les utiliser** mais elles sont pourtant très courantes, notamment sur les sites dont le code a été généré par un logiciel.

La balise ``<div>`` crée des **éléments de type *block*** qui se disposent les uns en dessous des autres, comme ``<p>``, ``<h1>``, ...

La balise ``<span>``, au contraire, crée des **éléments de type *inline*** qui sont insérés les uns à côté des autres (en ligne), comme ``<strong>``, ``<img>``, ``<a>``, ...

Avec le HTML5 sont arrivées de **nouvelles balises de structure**.

Par exemple, les balises ``<header>`` (à ne pas confondre avec ``<head>``) et ``<footer>`` créent aussi des éléments de type block mais réservés (idéalementà à l'**en-tête** (logo, banière, slogan, ...) et au **pied de page** (contact, nom, informations légales, ...) de la page web.

La balise ``<nav>`` crée un block dédié au **menu de navigation** sur le site sur la gauche de la page. Les balises ``<section>``, ``<article>`` et ``<aside>`` sont encore autant de solutions pour créer des blocs dans la page.

#### 4. Les commentaires

Comme pour un programme classique, il peut être utile de glisser quelques lignes de **commentaires** dans son code HTML.

On utilise pour cela la **balise** ``<!-- Ceci est un commentaire -->`` qui n'a pas besoin non plus d'être fermée.

#### 5. Mettre en valeur du texte

On a déjà vu la balise ``<strong>`` qui est une mise en valeur forte d'une portion de texte, traduite le plus souvent par des caractères **gras**.

Pour une mise en valeur moins forte, on peu utiliser la balise ``<em>``, pour *emphasize*, qui se traduit en général par de l'*italique*.

On peut aussi utiliser la balise ``<mark>`` pour mettre en évidence du texte dans un contexte donné. Cela aura pour effet de le surligner.

Evidemment, toutes ces balises doivent être fermées à la fin du morceau de texte concerné.

#### 6. Faire des listes

Pour démarrer une **liste à puces** (non numérotée) on utilise la balise ``<ul>`` (*unordered list*) et pour une **liste numérotée**, la balise ``<ol>`` (*ordered list*). Pour les terminer on utilise la **balise fermante associée**.

**Dans les deux cas**, chaque **élément de la list**e est délimité par les balises ``<li>`` et ``</li>`` (*list item*).

Exemple de **page structurée avec les balises précédentes** :

    <!doctype html>
    <html lang="fr">
        <head>
            <meta charset="utf-8">
            <title>Voici mon site</title>
        </head>
        <body>
            <!-- Ceci est un commentaire -->
            <div>div est une balise de type <strong>block</strong></div>
            <p><em>la balise p</em> est une autre balise de type block</p>
            <span>En revanche, span est une balise de type "inline"</span>
            <a href="www.google.fr">Et voici une autre balise de type "inline"</a>
            <h1>h1 est bien une balise de type "block"</h1>
            <span>la malheureuse balise span est <mark>obligée</mark> de se placer en dessous</span>
        </body>
    </html>

Voici ce que donne le code HTML précédent une fois interprété par le navigateur :
